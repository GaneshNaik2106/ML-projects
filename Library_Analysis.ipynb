```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

```

```python
data = pd.read_csv('Books_data - Sheet1.csv')
# Check the first few rows of the dataset
print(data.head())



```

```python
# Get a summary of the dataset
print(data.info())
```

```python
print(data.tail())
```

```python
# Compute summary statistics
print(data.describe())
```

```python
# Handle missing values
data.dropna(inplace=True)

# Convert 'status' column to categorical
data['status'] = data['status'].astype('category')
# Count plot for book categories
sns.countplot(x='category', data=data)
plt.title('Count of Books by Category')
plt.xticks(rotation=90)
plt.show()

```

```python
# Bar plot for book availability status
sns.countplot(x='status', data=data)
plt.title('Count of Books by Availability Status')
plt.show()

```

```python
# Histogram of book titles lengths
data['title_length'] = data['title'].apply(len)
plt.hist(data['title_length'], bins=20)
plt.title('Distribution of Title Lengths')
plt.xlabel('Title Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average title length by category
avg_title_length = data.groupby('category')['title_length'].mean()
print(avg_title_length)

# Proportion of books by availability status
status_proportion = data['status'].value_counts(normalize=True)
print(status_proportion)
```

```python

category_counts = data['category'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=180)
plt.title('Distribution of Books by Category')
plt.axis('equal') # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

```python
# Count Plot - Book Availability by Category
plt.figure(figsize=(10, 6))
sns.countplot(x='category', hue='status', data=data, palette='Set1')
plt.xticks(rotation=90)
plt.title('Availability of Books in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
df = pd.DataFrame(data)
# Perform basic data analysis
category_counts = df['category'].value_counts()
availability_counts = df['status'].value_counts()

# Plot the number of books per category
plt.figure(figsize=(12, 6))
plt.bar(category_counts.index, category_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Number of Books')
plt.title('Number of Books per Category')
plt.show()

# Plot the availability status of books
plt.figure(figsize=(6, 6))
plt.pie(availability_counts.values, labels=availability_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Availability Status')
```
